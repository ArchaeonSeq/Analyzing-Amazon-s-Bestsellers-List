# Question 1: The best selling books of the decade (2010 - 2019)

```python
bestsellers_list_copy = bestsellers_list.copy()
bestsellers_list_copy.drop(bestsellers_list_copy.loc[bestsellers_list_copy['Year']== 2009].index, inplace=True)
bestsellers_list_copy
```

```python
bestsellers_list_copy
```

```python
pre_bestselling_book1 = bestsellers_list_copy.groupby('Name')['Reviews'].sum()
pre_bestselling_book1 = pre_bestselling_book1.to_frame()
index = list(range(0,313))
pre_bestselling_book1['index'] = index

pre_bestselling_book1 = pre_bestselling_book1.set_index('index', append=True).reset_index(level=0)
pre_bestselling_book1
```

# Question 2: The trend of over the years

```python
for_trend = bestsellers_list.sort_values('Year' , 
                              ascending=True,
                              kind='heapsort', 
                              na_position='last')
for_trend
```

```python
Fiction = for_trend.loc[for_trend['Genre'] == 'Fiction']
Fiction
#x_axis
```

```python
Fiction_2009 = Fiction.loc[Fiction['Year'] == 2009]
a = Fiction_2009.Reviews.sum()

Fiction_2010 = Fiction.loc[Fiction['Year'] == 2010]
b = Fiction_2010.Reviews.sum()

Fiction_2011 = Fiction.loc[Fiction['Year'] == 2011]
c = Fiction_2011.Reviews.sum()

Fiction_2012 = Fiction.loc[Fiction['Year'] == 2012]
d = Fiction_2012.Reviews.sum()

Fiction_2013 = Fiction.loc[Fiction['Year'] == 2013]
e = Fiction_2013.Reviews.sum()

Fiction_2014 = Fiction.loc[Fiction['Year'] == 2014]
f = Fiction_2014.Reviews.sum()

Fiction_2015 = Fiction.loc[Fiction['Year'] == 2015]
g = Fiction_2015.Reviews.sum()

Fiction_2016 = Fiction.loc[Fiction['Year'] == 2016]
h = Fiction_2016.Reviews.sum()

Fiction_2017 = Fiction.loc[Fiction['Year'] == 2017]
i = Fiction_2017.Reviews.sum()

Fiction_2018 = Fiction.loc[Fiction['Year'] == 2018]
j = Fiction_2018.Reviews.sum()

Fiction_2019 = Fiction.loc[Fiction['Year'] == 2019]
k = Fiction_2019.Reviews.sum()

y_axis1 = [a, b, c, d, e, f, g, h, i, j, k]
y_axis1
```

```python
Non_Fiction = for_trend.loc[for_trend['Genre'] == 'Non Fiction']
Non_Fiction
```

```python
Non_Fiction_2009 = Non_Fiction.loc[Non_Fiction['Year'] == 2009]
l = Non_Fiction_2009.Reviews.sum()

Non_Fiction_2010 = Non_Fiction.loc[Non_Fiction['Year'] == 2010]
m = Non_Fiction_2010.Reviews.sum()

Non_Fiction_2011 = Non_Fiction.loc[Non_Fiction['Year'] == 2011]
n = Non_Fiction_2011.Reviews.sum()

Non_Fiction_2012 = Non_Fiction.loc[Non_Fiction['Year'] == 2012]
o = Non_Fiction_2012.Reviews.sum()

Non_Fiction_2013 = Non_Fiction.loc[Non_Fiction['Year'] == 2013]
p = Non_Fiction_2013.Reviews.sum()

Non_Fiction_2014 = Non_Fiction.loc[Non_Fiction['Year'] == 2014]
q = Non_Fiction_2014.Reviews.sum()

Non_Fiction_2015 = Non_Fiction.loc[Non_Fiction['Year'] == 2015]
r = Non_Fiction_2015.Reviews.sum()

Non_Fiction_2016 = Non_Fiction.loc[Non_Fiction['Year'] == 2016]
s = Non_Fiction_2016.Reviews.sum()

Non_Fiction_2017 = Non_Fiction.loc[Non_Fiction['Year'] == 2017]
t = Non_Fiction_2017.Reviews.sum()

Non_Fiction_2018 = Non_Fiction.loc[Non_Fiction['Year'] == 2018]
u = Non_Fiction_2018.Reviews.sum()

Non_Fiction_2019 = Non_Fiction.loc[Non_Fiction['Year'] == 2019]
v = Non_Fiction_2019.Reviews.sum()

y_axis2 = [l, m, n, o, p, q, r, s, t, u, v]
y_axis2
```

```python
for_x_axis = for_trend.Year
x_axis = for_x_axis.drop_duplicates(keep='first', inplace=False)
x_axis.reset_index(drop=True, inplace=True)
x_axis
```

```python
fig, ax = plt.subplots(figsize =(12, 8))
ax.set_xlabel("Year")
ax.set_ylabel("Sales")
ax.set_title("Trend Over The Years")
ax.plot(x_axis, y_axis1, marker = "o", linestyle = "-", color = "Yellow", label = "Fiction Sales")
ax.plot(x_axis, y_axis2, marker = "o", linestyle = "-", color = "Purple", label = "Non Fiction Sales")
plt.legend(loc="upper left")
plt.show()
```

This Line Chart shows a trend of over the years(2009 -2019) sales of fiction books and non fiction books. We can see from this chart that fiction generally had more sales than non-fiction from 2009 till 2017, then non-fiction started leading from 2018. 

We can also see that there was a general rise in fiction sales from 2009 till it reached its peak at 2014, then an overall decline from 2014 till 2018. Then a sharp increase from 2018 to 2019. 

We can also see that non-fiction sales generally had an increasing number of sales over the years with slight deviations in 2013 and 2017. We can also see that non-fiction sales reached its peak in 2018. 

```python
fig, ax = plt.subplots(figsize =(12, 8))
X = np.arange(11)
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, y_axis1, color = 'Yellow', width = 0.25,  label = "Fiction Sales")
ax.bar(X + 0.25, y_axis2, color = 'Purple', width = 0.25,  label = "Non Fiction Sales")
plt.legend(loc="upper left")
ax.set_xlabel("Year")
ax.set_ylabel("Sales")
ax.set_title("Trend Over The Years")
plt.xticks([r + 0.25 for r in range(11)], 
        ['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019'])
```

This Bar Chart shows a trend of over the years(2009 -2019) sales of fiction books and non fiction books. We can see from this chart that fiction generally had more sales than non-fiction from 2009 till 2017, then non-fiction started leading from 2018. 

We can also see that there was a general rise in fiction sales from 2009 till it reached its peak at 2014, then an overall decline from 2014 till 2018. Then a sharp increase from 2018 to 2019. 

We can also see that non-fiction sales generally had an increasing number of sales over the years with slight deviations in 2013 and 2017. We can also see that non-fiction sales reached its peak in 2018.
